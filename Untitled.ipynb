```python
import matplotlib.pyplot as plt

# Artery names and accuracies extracted from the graph for group A
artery_names = ['ACA', 'ICA', 'MCA', 'PCA']
accuracies = [98.88, 92.75, 87.94, 91.23]

# Define specific colors for each artery
colors = ['#3CAEA3',  # Green
          '#20639B',  # Blue
          '#173F5F',  # Dark Blue
          '#937AFF']  # Light Purple

# Use a clean style
plt.style.use('seaborn-poster')

# Create a figure and axis for the plot
fig, ax = plt.subplots(figsize=(10, 6))

# Create a bar plot for group A with individual colors for each bar
bars = ax.bar(artery_names, accuracies, color=colors)

# Annotate the bars with their respective accuracy values
for i, value in enumerate(accuracies):
    ax.text(i, value + 1, f"{value:.2f}%", ha='center', va='bottom', fontsize=14, color='black', fontweight='bold')

# Customize the plot
ax.set_title('Accuracy of Stenosis Prediction by Arteries (98 TOF-MRAs)', fontsize=18, pad=20, fontweight='bold')
ax.set_xlabel('Artery', fontsize=16, labelpad=10)
ax.set_ylabel('Accuracy (%)', fontsize=16, labelpad=10)

# Add gridlines for readability
ax.grid(True, which='both', axis='y', linestyle='--', linewidth=0.7, alpha=0.7)

# Increase the font size for the x and y axis ticks
ax.tick_params(axis='x', labelsize=14)
ax.tick_params(axis='y', labelsize=14)

# Set y-axis limits slightly above 100 for better visualization of annotations
ax.set_ylim(0, 105)

# Remove unnecessary spines to make it look cleaner
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Add a background color to make the plot stand out
fig.patch.set_facecolor('#f9f9f9')

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Show the plot
plt.tight_layout()
plt.show()
```
